# Merton model and Poisson process with Log Normal intensity function

## 参考

### ロジスティック関数とプロビット関数の近似

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# xの範囲を設定
x = np.linspace(-5, 5, 100)

# 標準正規分布の分布関数のyの値を計算
y1 = norm.pdf(x, 0, 1)


# ビショップを参考にしたbeta = 1.596
y3 = 1.596 / (np.exp(1.596 * x) * (1 + np.exp(-1.596 * x))**2)

# 論文指定のbeta = 1.3
y5 = 1.3 / (np.exp(1.3 * x) * (1 + np.exp(-1.3 * x))**2)

# プロット
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Standard Normal Distribution Function')
plt.plot(x, y3, label='Derivative of 1 / (1 + exp(-1.596x))')
plt.plot(x, y5, label='Derivative of 1 / (1 + exp(-1.3x))', linestyle='--')
plt.title('Function Plots')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

### 二項分布とポアソン分布の近似

```python
import numpy as np
from scipy.stats import binom, poisson
import matplotlib.pyplot as plt

# xの範囲を設定
x = np.arange(0, 21)

# 二項分布の確率質量関数のyの値を計算
y1 = binom.pmf(x, 20, 0.1)

# ポアソン分布の確率質量関数のyの値を計算
y2 = poisson.pmf(x, 2)

# プロット
plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'o-', label='Binomial (n=20, p=0.1)')
plt.plot(x, y2, 'o-', label='Poisson (lambda=2)')
plt.title('Binomial and Poisson Distributions')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

### 対数正規分布

```python
import numpy as np
from scipy.stats import lognorm
import matplotlib.pyplot as plt

# xの範囲を設定
x = np.linspace(0.01, 5, 100)

# yの値を計算（標準対数正規分布の確率密度関数）
y = lognorm.pdf(x, 1)

# プロット
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Log-normal Distribution Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
